```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('walmart.csv')
df.head(10)
```

```python
r1 = df['Temperature'].value_counts().reset_index()
r1.head()
```

```python
sns.barplot(data = r1,x ='Temperature',ci= None)
plt.title(" Temperature plot ")
plt.show()
```

```python
r2 = df.groupby('Date')[['Temperature','Fuel_Price']].mean()
r2.head()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.barplot(data = r2,x ='Temperature',y ='Fuel_Price',ci= None)
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.lineplot(data = r2,x ='Temperature',y ='Fuel_Price',marker ='d',ci= None)
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.scatterplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
#sns.set(rc={'figure.figsize':(9,4)})
sns.boxplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.violinplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.stripplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.swarmplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.displot(data = r2,x ='Temperature',y ='Fuel_Price',kind ='hist')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.countplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.set_style('darkgrid')
sns.set_context('poster',font_scale=0.7,rc ={'grid.linewidth':0.8})
sns.barplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(9,4)})
sns.set_style('darkgrid')
sns.set_context('poster',font_scale=0.8,rc ={'grid.linewidth':0.5})
sns.scatterplot(data = r2,x ='Temperature',y ='Fuel_Price')
plt.title(" Temperature and Fuel_Price plot ")
plt.show()
```
