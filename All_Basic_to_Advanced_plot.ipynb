```python
# Data visualization
```

##### Barplot

Definition: A bar plot is used to compare relationshiop between categorical variable and numerical variable
eg. Comparing sales across different product categories 


```python
data['Category'].value_counts()
```

```python
a1 = data.groupby('Category')[['Sales']].sum().reset_index()
```

```python
a1
```

```python
plt.bar(a1['Category'], a1['Sales'],edgecolor='black')
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=90)
plt.show()
```

#### Scatter plot

Scatter plot used to show relationship between two numerical variable

```python
plt.scatter(data['Quantity'], data['Discount'],color='green')
plt.title('Quantity vs Discount')
plt.xlabel('Quantity')
plt.grid()
plt.ylabel('Discount')
plt.show()
```

### Line Plot
A line plot is used to Show time series data
eg.sales by week, month, or year.

```python
monthly_sales = data.groupby(data['Ship Date'].dt.month) ['Sales'].sum().reset_index()
```

```python
monthly_sales
```

```python
plt.plot(monthly_sales['Ship Date'].astype(str), monthly_sales['Sales'], marker='o',color='red')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.grid()
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=90)
plt.show()
```

#### Histogram

Definition: A histogram is used to visualize the distribution of a single variable by dividing data into bins and counting how many data points fall into each bin.
eg. To analyze the distribution of sales amounts, customer counts

```python
plt.hist(data['Sales'], bins=20, color='Orange', edgecolor='black')
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales ($)')
plt.grid()
plt.ylabel('Frequency')
plt.show()
```

#### Advanced Plot

##### Heatmap
To visulize the correlation Between multiple variable

```python
corr = data[['Sales', 'Quantity', 'Discount','Profit', 'Shipping Cost']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

##### Pairplot
Definition: A pair plot creates scatter plots for each pair of features, allowing for the exploration of relationships between multiple numerical variables.

```python
sns.pairplot(data,vars=['Sales', 'Quantity', 'Discount','Profit', 'Shipping Cost'])
```

##### Violinplot

Defination :- A violin plot combines aspects of box plots and density plots.
It shows the distribution of data across different categories and and is particularly useful for comparing distributions.



```python
data['Category'].value_counts()
```

```python
sns.violinplot(x='Category', y='Sales', data=data)
plt.title('Sales Distribution by Category')
plt.grid()
plt.show()
```

###### Density Plot (Kernal Density Estimate) KDE plot

A density plot is a smoothed version of the histogram, showing the probability distribution of a continuous variable.

```python
sns.kdeplot(data['Sales'],shade= True,color='blue')
plt.title('Density Plot of Total Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Density')
plt.show()
```

###### Stacked Bar chart
Definition: A stacked bar plot shows the values of different categories stacked on top of one another within a single bar.

```python
a3 = pd.pivot_table(data= data, index=['Category'],values=['Sales'],aggfunc=sum)
a3
```

```python
a3.index
```

```python
a3.Sales
```

```python
a3.plot(kind='bar',stacked= True, figsize=(5,5),color='Yellow',edgecolor='black')
plt.title("Stacked bar chart")
plt.show()
```

###### Bubble chart
Definition: A bubble chart is a variation of a scatter plot where each point is represented as a bubble
It is useful for visualizing relationships between three continuous variables.

```python
plt.scatter(data['Quantity'], data['Sales'], s=data['Profit']/5, alpha=0.5, color='blue')
plt.title('Bubble Chart: Quantiy vs. Total Sales')
plt.xlabel('Quantity')
plt.ylabel('Total Sales ($)')
plt.show()
```

##### Area Plot
Definition: An area plot is similar to a line plot, but the area under the line is filled with color. 
It is helpful for visualizing trends over time 

```python
monthly_sales = data.groupby([data['Ship Date'].dt.to_period('M'),'Region'])['Sales'].sum().unstack()
```

```python
monthly_sales.plot(kind='area',figsize=(10,10))
plt.show()
```

###### Donut Chart
A donut chart is essentially a pie chart with a hole in the middle. It is used to represent parts of a whole, where each slice shows a proportion of the total. 


```python
a4 = data.groupby('Region')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False).head(5)
a4
```

```python
plt.pie(a4['Sales'],labels= a4['Region'],autopct="%.2f%%",shadow= True, wedgeprops={'width':0.4})
plt.legend(loc='upper left',fontsize=8,bbox_to_anchor=(1.1,1.1))
plt.show()
```
